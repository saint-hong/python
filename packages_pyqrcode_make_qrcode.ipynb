### kream의 스투시 반바지 판매 페이지 QR Code 생성

```python
import pyqrcode
from PIL import Image
```

```python
link = input("Enter anyting to generate QR : ")
qr_code = pyqrcode.create(link)
qr_code.png("QRCode.png", scale=5)
Image.open("QRCode.png")
```
