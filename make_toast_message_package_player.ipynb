## Toast Message 알람 띄우기

```python
from plyer import notification
```

```python
notification.notify(
    title="Notice!",
    message="Hello, this is a toast message.",
    app_name="Useful Python",
    app_icon=None,
    timeout=10
)

if __name__ == "__main__" : 
    notification
```

### 함수와 함께 사용
- x + y의 결과값을 메시지로 담은 알림을 울릴 수 있다. 

```python
def calc(x, y) : 
    
    result = x + y
    
    return notification.notify(
            title="Finish Func", 
            message=f"func result : {result}", 
            app_name="func checker", 
            app_icon=None, 
            timeout=30,
            toast=True)

if __name__ == "__main__" : 
    calc(5, 5)
```
